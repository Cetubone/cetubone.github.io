## 3-feruloylquinic acid

```python
import pandas as pd
file_path = r"C:\Users\Administrator\Desktop\SO20250331-138&139&141\excels\SPR-3_feruloylquinic_acid\RU-Time.xlsx"
RU_Time_df = pd.read_excel(file_path)
RU_Time_df = RU_Time_df.iloc[:2885]

tmp_df = RU_Time_df[RU_Time_df["Time"]>=0]

RU_Time_0_df = RU_Time_df[RU_Time_df["Time"]<0]
RU_Time_0_85_df = tmp_df[tmp_df["Time"]<85]
RU_Time_85_df = tmp_df[tmp_df["Time"]>=85]

RU_Time_df.columns
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))

colInfo_dic = {
    "1.56": "#C64A4A",
    "3.13": "#1FB934",
    "6.25": "#3B43D4",
    "12.5": "#6A108E",
    "25.0": "#99931E",
    "50.0": "#E87F35"
}

for col in RU_Time_0_df.columns.tolist()[2:][::-1]:
    plt.plot(
        RU_Time_0_df["Time"],
        RU_Time_0_df[col],
        color=colInfo_dic[col], label=col + " mol")

for col in RU_Time_0_85_df.columns.tolist()[2:][::-1]:
    plt.plot(
        RU_Time_0_85_df["Time"],
        RU_Time_0_85_df[col],
        color=colInfo_dic[col])

for col in RU_Time_85_df.columns.tolist()[2:][::-1]:
    plt.plot(
        RU_Time_85_df["Time"],
        RU_Time_85_df[col],
        color=colInfo_dic[col])

plt.xlabel('X')
plt.ylabel('Y')
plt.title('XXX')
plt.legend()
plt.tight_layout()
plt.show()
```
